```python
import numpy as np 

A = np.array([0.0, 1.0])  # 出発点 (固定)
B = np.array([1.0, 0.0])  # 終点 (固定)

n = 10  # ユーザーが指定した制御点の数
x_range = np.linspace(0, 1, n)  # x 座標は等間隔
y_init = np.linspace(A[1], B[1], n)  # 初期y座標（直線）
y_init[0], y_init[-1] = A[1], B[1]  # A, B は固定

y_data = (x_range-1)**2

print(x_range)
print(y_data)
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib

fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(x_range, y_data, '.', label="初期曲線")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
ax.grid()

fig.show()
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib

x_data_g = np.gradient(x_range)
y_data_g = np.gradient(y_data)
fig, ax = plt.subplots(figsize=(6, 6))
print(x_data_g)
print(y_data_g[1:9])
ax.plot(x_data_g[1:9], y_data_g[1:9], '.', label="初期曲線")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
ax.grid()

fig.show()
```
